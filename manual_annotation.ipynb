```python
import pandas as pd

# Code generated using Claude Sonnet 4, with the following prompt:
# "i have a df with this format:
# PRODUCT_IDTITLEATTRIBUTESSOURCE_FILE0MLB29523289Base Estrela Secretaria Em A√ßo C/ Rod√≠zios E C...{'Marca': 'cadeira girat√≥ria diretora', 'Fabri...info_products_10_escrit√≥rio.json1MLB34230359Escrivaninha Office Nt2105 Not√°vel M√≥veis Cor ...{'Marca': 'Not√°vel M√≥veis', 'Modelo': 'NT2105'...info_products_10_escrit√≥rio.json2MLB15904595Cadeira de Escrit√≥rio Mymax Gamer MX5 Cor Preto{'Marca': 'Mymax', 'Linha': 'Gamer', 'Modelo':...info_products_10_escrit√≥rio.json3MLB26801865Marmita Dupla Camada √Å Prova De Vazamento Imed...{'Marca': 'Marmita,Lancheira,Escritorio', 'Mod...info_products_10_escrit√≥rio.json4MLB22522670Cadeira de escrit√≥rio TGTTGT-OFC-001 TGT ergon...{'Marca': 'TGT', 'Linha': 'cadeira fixa escrit...info_products_10_escrit√≥rio.json
# I want to create a column "IS_HAZMAT_HUMAN" and for at least 15 lines for each source_file i want to fill it manually in a easy way using python"""

def label_hazmat_products(df, samples_per_file=15):
    """
    Interactive labeling tool for hazmat classification.
    Shows products and lets you mark them as hazmat (y/n).
    """
    
    # Create the new column with default None
    df['IS_HAZMAT_HUMAN'] = None
    
    # Get unique source files
    source_files = df['SOURCE_FILE'].unique()
    
    print(f"Found {len(source_files)} source files:")
    for i, file in enumerate(source_files):
        print(f"{i+1}. {file}")
    
    print(f"\nWill label {samples_per_file} products from each file.")
    print("Instructions:")
    print("- Type 'y' for hazmat products")
    print("- Type 'n' for non-hazmat products") 
    print("- Type 's' to skip")
    print("- Type 'q' to quit")
    print("- Type 'save' to save progress\n")
    
    total_labeled = 0
    
    for source_file in source_files:
        print(f"\n{'='*60}")
        print(f"SOURCE FILE: {source_file}")
        print(f"{'='*60}")
        
        # Get products from this source file
        file_products = df[df['SOURCE_FILE'] == source_file].head(samples_per_file)
        
        for idx, row in file_products.iterrows():
            print(f"\n--- Product {idx} ---")
            print(f"TITLE: {row['TITLE']}")
            
            # Show attributes if they exist
            if pd.notna(row['ATTRIBUTES']):
                print(f"ATTRIBUTES: {str(row['ATTRIBUTES'])[:200]}...")
            
            while True:
                response = input(f"\nIs this hazmat? (y/n/s/q/save): ").lower().strip()
                
                if response == 'q':
                    print(f"\nQuitting. Labeled {total_labeled} products total.")
                    return df
                elif response == 'save':
                    print(f"Saving progress... ({total_labeled} products labeled)")
                    return df
                elif response == 'y':
                    df.loc[idx, 'IS_HAZMAT_HUMAN'] = True
                    total_labeled += 1
                    print("‚úì Marked as HAZMAT")
                    break
                elif response == 'n':
                    df.loc[idx, 'IS_HAZMAT_HUMAN'] = False
                    total_labeled += 1
                    print("‚úì Marked as NOT HAZMAT")
                    break
                elif response == 's':
                    print("Skipped")
                    break
                else:
                    print("Please enter 'y', 'n', 's', 'q', or 'save'")
        
        print(f"\nCompleted {source_file} - {total_labeled} products labeled so far")
    
    print(f"\nüéâ All done! Labeled {total_labeled} products total.")
    return df

    

print("Sample usage:")
print("df = label_hazmat_products(df)")
print("\nReplace the sample_data with your actual dataframe loading code.")

# Call the labeling function
df = label_hazmat_products(df)
```

```python
# Save only labeled products to a new CSV file
labeled_df = df[df['IS_HAZMAT_HUMAN'].notna()]
labeled_df.to_csv('data/dataset_1/dataset_1_annotated.csv', index=False)
```
